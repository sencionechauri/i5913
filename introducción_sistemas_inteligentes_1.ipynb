## <center>¿Qué son los sistemas basados en conocimiento?</center>

El área de los sistemas basados en conocimiento es una aproximación muy exitosa al problema clásico de la Inteligencia Artificial de programar inteligencia.

## <center>Una definición</center>

Un programa inteligente de computadora que utiliza **conocimiento** y procedimientos de **inferencia** para resolver problemas que son suficientemente difíciles como para requerir **conocimiento experto humano** para sus soluciones. (Feigenbaum)

## <center>Emular, no imitar</center>

Un sistema experto es un sistema computacional que emula la habilidad de toma de decisiones de un experto humano.

## <center>Una rama</center>

Los sistemas expertos son una rama de la inteligencia artificial que hace uso extensivo de **conocimiento especializado** para resolver problemas a nivel de un **experto** humano.

## <center>Nombres comunes</center>

* sistemas expertos
* sistemas basados en conocimiento
* sistemas expertos basados en conocimiento

## <center>Dominio de un problema</center>

El conocimiento de un experto es específico al Dominio de problema.

El _Dominio del problema_ es el área especial del problema tal como medicina, finanzas, ciencia, ingeniería, etc. para la cual un experto puede resolver problemas muy bien.

## <center>Dominio de conocimiento</center>

El conocimiento del experto sobre cómo resolver problemas específicos es llamado el Dominio de conocimiento del experto.

## <center>Relación entre Dominio de problema y Dominio de conocimiento</center>

<center><img src="https://raw.githubusercontent.com/sencionechauri/i5913/master/media/problema_conocimiento.png" width=75% /></center>

Dentro del dominio de conocimieto del cual sabe, el sistema experto **razona** o hace **inferencias** de la misma forma que un humano inferiría la solución a un problema.
